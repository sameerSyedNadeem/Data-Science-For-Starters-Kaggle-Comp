```python
X=cleaned_df.drop(columns=[target, 'row ID'])
y=cleaned_df[target]

y
```

```python
corCheck = X.corr()


# Set a threshold for correlation coefficient
threshold = 0.7

# Drop features that have a correlation coefficient higher than the threshold
to_drop = []
for i in range(len(corCheck.columns)):
    for j in range(i):
        if abs(corCheck.iloc[i, j]) > threshold:
            colname = corCheck.columns[i]
            to_drop.append(colname)
            
X_new = X.drop(to_drop, axis=1)
```

FIT MULTIPLE MODELS TO DECIDE THE BEST ONE 
(*criterion: Least RMSE*)

```python
def fitting_models():
    lr=LinearRegression()
    ls = Lasso(alpha=2)
    rg = Ridge(alpha=8)
    gbr = GradientBoostingRegressor(random_state = 69, n_estimators = 1000)
    #br = BaggingRegressor(n_estimators = 26, random_state = 5)
    
    rgs = [('Linear Regression', lr),
        ('Lasso', ls),
        ('Ridge', rg),
      #  ('Bagging Regressor', br)
        ('GBR', gbr)
              
  ]
    for name,lr in rgs:
        lr.fit(X_train,y_train)
        pred = lr.predict(X_test)
        score = format(r2_score(y_test,pred), '.4f')
        rmse = np.sqrt(mean_squared_error(y_test, pred))
        print("{} : {} , {}".format(name,score,rmse))
```
