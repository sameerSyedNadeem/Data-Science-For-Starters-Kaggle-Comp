DEFINE X/y

```python
cleaned_df = data
```

```python
target = "loss"
```

```python
X=cleaned_df.drop(columns=[target,'f1','f3','f5','f9','f6'])
y=cleaned_df[target]
y
```

LABEL ENCODE CATEGORICAL VARIABLES
*(You can also use one-hot encoding)*

```python
cat_cols = X.select_dtypes(include=['object']).columns
le = LabelEncoder()
X = X[cat_cols].apply(lambda x: le.fit_transform(x))
```

```python
X=X.drop(columns=['cat98','cat3','cat7'])
```

TRAIN/VALIDATION SPLIT

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

FIT MULTIPLE MODELS TO DECIDE THE BEST ONE 
(*criterion: Least RMSE*)

```python
def fitting_models():
    lr=LinearRegression()
    ls = Lasso(alpha=5)
    rg = Ridge(alpha=20)
    
    rgs = [('Linear Regression', lr),
        ('Lasso', ls),
        ('Ridge', rg)
              
  ]
    for name,lr in rgs:
        lr.fit(X,y)
        pred = lr.predict(X)
        score = format(r2_score(y,pred), '.4f')
        rmse = np.sqrt(mean_squared_error(y, pred))
        print("{} : {} , {}".format(name,score,rmse))
```

```python
fitting_models()
```

```python
lr=LinearRegression().fit(X_train, y_train)
```

```python
pred = lr.predict(X)
score = format(r2_score(y,pred), '.4f')
rmse = np.sqrt(mean_squared_error(y, pred))
print("{} , {}".format(score,rmse))
```
